## Ejercicio 2

```python
def asleep(N):
    if N == 0:
        return 'INSOMNIA'
    
    unique_digits = {int(d) for d in str(N)}
    zero_to_nine = set()
    zero_to_nine.update(unique_digits)
    i=2
    
    while(len(zero_to_nine) != 10):
        new_N = N*i
        unique_digits = {int(d) for d in str(N*i)}
        zero_to_nine.update(unique_digits)
        i+=1
        
    return new_N

asleep(1692)
```
